```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
```

```python
target_train = pd.read_csv('Data/train.csv')
target_train.head()
```

```python
book_train = pd.read_parquet('Data/book_train.parquet/stock_id=0')
book_train['stock_id'] = 0
book_train['wap1'] = (book_train['bid_price1'] * book_train['ask_size1'] + book_train['ask_price1'] * book_train['bid_size1'])/(book_train['bid_size1']+book_train['ask_size1'])
book_train['wap2'] = (book_train['bid_price2'] * book_train['ask_size2'] + book_train['ask_price2'] * book_train['bid_size2'])/(book_train['bid_size2']+book_train['ask_size2'])
book_train = book_train[book_train['time_id']==5]
book_train.head()
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=book_train['seconds_in_bucket'], y=book_train['wap1'], mode='lines+markers', name='WAP1'))
fig.add_trace(go.Scatter(x=book_train['seconds_in_bucket'], y=book_train['wap2'], mode='lines+markers', name='WAP2'))
fig.show()
```

```python
def log_return(list_stock_prices):
    return np.log(list_stock_prices).diff()
```

```python
book_train['log_return1'] = log_return(book_train['wap1'])
book_train['log_return2'] = log_return(book_train['wap2'])
book_train.head()
```

```python
book_train.dropna(subset = ['log_return1','log_return2'], inplace=True)
book_train.head()
```

```python
def realized_volatility(series_log_returns):
    return np.sqrt(np.sum(series_log_returns**2))
```

```python
realized_vol1 = realized_volatility(book_train['log_return1'])
print('Realized volatility (1) for stock_id=0 and time_id=5 is', realized_vol1)
realized_vol2 = realized_volatility(book_train['log_return2'])
print('Realized volatility (2) for stock_id=0 and time_id=5 is', realized_vol2)
```
